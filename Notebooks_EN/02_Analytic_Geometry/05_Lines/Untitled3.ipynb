* The line passes through points $A(1, 2)$ and $B(3, 4)$. Find the equation of the line.
* The line passes through point $A(1, 2)$ and is parallel to the line $y = 2x + 3$. Find the equation of the line.

* Calculate the angle between the line $y = x + 3$ and the $Ox$ axis.
* Provide a vector perpendicular to the line $x + y + 1 = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Points A and B
A = (1, 2)
B = (3, 4)

# Line equation parameters
x = np.linspace(0, 4, 100)  # x values
y = x + 1  # Line equation y = x + 1

# Plot the line
plt.plot(x, y, label="Line: y = x + 1", color="blue")

# Plot points A and B
plt.scatter(*A, color="red", label="Point A (1, 2)")
plt.scatter(*B, color="green", label="Point B (3, 4)")

# Formatting the graph
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.title("Line through Points A and B")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0, 4)
plt.ylim(0, 5)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equations of the lines
x = np.linspace(-1, 3, 100)
y1 = 2*x + 3  # Line 1
y2 = 3*x + 2  # Line 2

# Intersection point
intersection_x = 1
intersection_y = 5

# Calculate the angle
m1 = 2
m2 = 3
theta_radians = np.arctan(abs((m2 - m1) / (1 + m1*m2)))
theta_degrees = np.degrees(theta_radians)

# Plot the lines
plt.plot(x, y1, label="Line 1: y = 2x + 3", color="blue")
plt.plot(x, y2, label="Line 2: y = 3x + 2", color="red")

# Plot the intersection point
plt.scatter(intersection_x, intersection_y, color="black", label="Intersection Point (1, 5)")

# Formatting the graph
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.title(f"Intersection and Angle (  {theta_degrees:.2f})")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-1, 3)
plt.ylim(-1, 10)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define line and point
m, c = 2, 3  # Line: y = 2x + 3
x1, y1 = 1, 2  # Point A(1, 2)

# Calculate distance
numerator = abs(m * x1 - y1 + c)
denominator = np.sqrt(m**2 + 1)
distance = numerator / denominator

print(f"Distance from A(1, 2) to the line y = 2x + 3: {distance:.3f}")

# Plotting
x = np.linspace(-1, 3, 100)
y = m*x + c  # Line equation
plt.plot(x, y, label="Line: y = 2x + 3", color="blue")
plt.scatter(x1, y1, color="red", label="Point A(1, 2)")

# Perpendicular projection
x_perp = (x1 + m*y1 - m*c) / (m**2 + 1)
y_perp = (m*x_perp + c)
plt.plot([x1, x_perp], [y1, y_perp], 'k--', label="Perpendicular Distance")

# Formatting
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.title("Distance from a Point to a Line")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-1, 3)
plt.ylim(-1, 8)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the coefficients from the line equation
a = 1  # Coefficient of x
b = 1  # Coefficient of y

# The normal vector to the line (perpendicular vector)
normal_vector = np.array([a, b])

# Define the line equation: y = -x - 1 (Rearranged form of x + y + 1 = 0)
x_vals = np.linspace(-10, 10, 100)
y_vals = -x_vals - 1

# Create the plot
plt.figure(figsize=(6,6))

# Plot the line
plt.plot(x_vals, y_vals, label="Line: x + y + 1 = 0", color="blue")

# Plot the normal vector (perpendicular vector)
origin = np.array([0, 0])  # Starting point of the normal vector
plt.quiver(origin[0], origin[1], normal_vector[0], normal_vector[1], angles='xy', scale_units='xy', scale=1, color="red", label="Normal Vector")

# Set the plot limits and labels
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Line and Perpendicular Vector')
plt.show()
```
